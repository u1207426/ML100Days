```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間
參考 Day 17 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
# 先看看離散化前AGES、DAYS_BIRTH與TARGET的相關係數
app_train["AGES"] = app_train["DAYS_BIRTH"]/365
print("DAYS_BIRTH與TARGET的相關係數： ", app_train["DAYS_BIRTH"].corr(app_train["TARGET"]))
print("AGES與TARGET的相關係數： ", app_train["AGES"].corr(app_train["TARGET"]))
app_train["AGES"].describe()
```

```python
#先把幾個重點欄位TARGET、AGES、DAYS_BIRTH拉出來另存成一個dataframe
age_data = app_train[["TARGET", "DAYS_BIRTH", "AGES"]]

#對AGES欄位進行分群存在新增的EQUAL_WIDTH_AGES欄位中
age_data["EQUAL_WIDTH_AGES"] = pd.cut(age_data["AGES"], bins=np.linspace(20, 70, num=11))
#age_data[["DAYS_BIRTH", "AGES", "EQUAL_WIDTH_AGES", "TARGET"]].head(50)
age_data.loc[:50, ["DAYS_BIRTH", "AGES", "EQUAL_WIDTH_AGES", "TARGET"]]
```

```python
#計算每個年齡區間的 Target、DAYS_BIRTH與AGES的平均值，這邊的重點是在看分群後EQUAL_WIDTH_AGES與TARGET是不是有相關性
age_groups = age_data.groupby("EQUAL_WIDTH_AGES").mean()
age_groups
```

```python
plt.figure(figsize = (8, 8))

px = age_groups.index.astype(str)
py = age_groups["TARGET"]*100
plt.bar(px, py)

plt.xticks(rotation = 75); plt.title('Failure to Repay by Age Group')
plt.xlabel("Age Group(Years)"); plt.ylabel("Average Failure to Repay")
plt.show()
```

可以非常明顯地看出趨勢，越年輕的客戶族群的違約遲繳的可能越大！
